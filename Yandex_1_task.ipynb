```python
import numpy
import math  
import random 
import matplotlib.pyplot as plt
def generate1(): 
    a = random.uniform(0, 1) 
    b = random.uniform(0, 1) 
    return (a * math.cos(2 * math.pi * b), a * math.sin(2 * math.pi * b)) 
ax = []
ay = []
for i in range(0, 250): 
    x, y = generate1() 
    ax.append(x) 
    ay.append(y) 
plt.scatter(ax, ay, color='red')  
plt.show()
```

```python
def generate2(): 
    while True: 
        x = random.uniform(-1, 1) 
        y = random.uniform(-1, 1)
        if x**2 + y**2 > 1: 
            continue 
        return (x, y) 
bx = [] 
by = []
for i in range(0, 250): 
    x, y = generate2()
    bx.append(x) 
    by.append(y) 
plt.scatter(bx, by, color='purple') 
plt.show()
```
